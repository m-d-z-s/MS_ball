```python
from math import cos, sin

import matplotlib.pyplot as plt
g = 9.8 # used for gravity constant
```

```python
v0 = 1.0
dt = 0.00001
x0 = 0.0
y0 = 0.0
alpha = 0.8
```

```python
vx0 = v0 * cos(alpha)
vy0 = v0 * sin(alpha)
```

```python
# vyn = vyn-1 - g * dt
# vxn = vxn-1
# yn = yn-1 + vyn * dt
# xn = xn-1 + vxn * dt
```

```python
def get_next_state(previous_state: tuple) -> tuple:
    x, y, vx, vy = previous_state
    vy_next = vy - g * dt
    if y < 0:
        vy_next *= -1
    y_next = y + vy_next * dt
    x_next = x + vx * dt
    return x_next, y_next, vx, vy_next
```

```python
def get_trajectory() -> list:
    trajectory = []
    state = (x0, y0, vx0, vy0)
    for i in range(100000):
        trajectory.append(state)
        state = get_next_state(state)
    return trajectory


trajectory = get_trajectory()
```

```python
def plot_trajectory(trajectory: list):
    x = [state[0] for state in trajectory]
    y = [state[1] for state in trajectory]
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axhline(y=0, color='green', linestyle='-')
    plt.show()


plot_trajectory(trajectory)
```
